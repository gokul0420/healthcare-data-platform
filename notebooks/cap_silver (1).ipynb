```python
#Check Missing Values
from pyspark.sql.functions import col, sum

def check_nulls(df, name):
    print(f"\nMissing values in {name}:")
    df.select([sum(col(c).isNull().cast("int")).alias(c) for c in df.columns]).show()

check_nulls(df_adm, "df_adm")
check_nulls(df_pat, "df_pat")
check_nulls(df_doc, "df_doc")
check_nulls(df_hosp, "df_hosp")
check_nulls(df_bill, "df_bill")
check_nulls(df_lab, "df_lab")
```

```python
#Check Duplicates
from pyspark.sql.functions import col

def check_duplicates(df, name):
    print(f"\nDuplicates in {name}:")
    
    total_count = df.count()
    distinct_count = df.dropDuplicates().count()
    
    print(f"Total Rows     : {total_count}")
    print(f"Distinct Rows  : {distinct_count}")
    print(f"Duplicate Rows : {total_count - distinct_count}")

check_duplicates(df_adm, "df_adm")
check_duplicates(df_lab, "df_lab")
check_duplicates(df_pat, "df_pat")
check_duplicates(df_doc, "df_doc")
check_duplicates(df_hosp, "df_hosp")
check_duplicates(df_bill, "df_bill")
```

Derivation of columns

```python
from pyspark.sql.functions import datediff

df_adm = df_adm.withColumn(
    "LOS",
    datediff(col("discharge_date"), col("admission_date"))
)
```

```python
df_adm = df_adm.withColumn(
    "risk_flag",
    (col("LOS") > 10).cast("int")
)
```

Re admission flag

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import col, lead, datediff

window_spec = Window.partitionBy("patient_id") \
                    .orderBy("admission_date")
```

```python
df_adm = df_adm.withColumn(
    "next_admission_date",
    lead("admission_date").over(window_spec)
)
```

```python
df_adm = df_adm.withColumn(
    "days_to_next_admission",
    datediff(col("next_admission_date"), col("discharge_date"))
)
```

```python
from pyspark.sql.functions import when

df_adm = df_adm.withColumn(
    "readmission_flag",
    when((col("days_to_next_admission") <= 30) & (col("days_to_next_admission") >= 0), 1).otherwise(0)
)
```

```python
df_adm = df_adm.drop("next_admission_date", "days_to_next_admission")
```

```python
df_adm.select("patient_id", "admission_date", "discharge_date", "readmission_flag").display()
```
