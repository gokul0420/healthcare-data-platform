```python


from pyspark.sql.functions import col

df_adm = spark.table("capstone.silver.admissions")
df_pat = spark.table("capstone.silver.patients")
df_bill = spark.table("capstone.silver.billing")
df_lab = spark.table("capstone.silver.lab_results")
```

```python

#valid
df_invalid_adm = df_adm.filter(
    col("discharge_date") < col("admission_date")
)

df_valid_adm = df_adm.filter(
    col("discharge_date") >= col("admission_date")
)
```

```python
df_invalid_adm.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("capstone.validation.invalid_admissions")
```

```python
# =========================================
# VALIDATION: PATIENT
# =========================================

df_invalid_pat = df_pat.filter(col("age") <= 0)

df_valid_pat = df_pat.filter(col("age") > 0)
```

```python
df_invalid_pat.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("capstone.validation.invalid_patient")
```

```python
# =========================================
# VALIDATION: BILLING
# =========================================

df_invalid_bill = df_bill.filter(
    col("total_charges") < col("insurance_covered_amount")
)

df_valid_bill = df_bill.filter(
    col("total_charges") >= col("insurance_covered_amount")
)
```

```python
df_invalid_bill.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("capstone.validation.invalid_billing")
```

```python
# =========================================
# VALIDATION: LAB
# =========================================

df_invalid_lab = df_lab.filter(
    col("normal_range_flag").isNull()
)

df_valid_lab = df_lab.filter(
    col("normal_range_flag").isNotNull()
)
```

```python
df_invalid_lab.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("capstone.validation.invalid_lab")
```

```python

print("Invalid Admissions :", df_invalid_adm.count())
print("Invalid Patients   :", df_invalid_pat.count())
print("Invalid Billing    :", df_invalid_bill.count())
print("Invalid Lab        :", df_invalid_lab.count())
```
